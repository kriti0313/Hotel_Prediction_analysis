# sorting data according to month

```python
from calendar import month_name

def sort_month(dataset,colname):
    month_dict = { j:i for i,j in enumerate(month_name)} #dictionary comprehension
    dataset["month_num"] = dataset[colname].apply(lambda x : month_dict[x])
    return dataset.sort_values(by="month_num").reset_index().drop(['index','month_num'],axis=1)
```

# Feature Encoding 

# Mean Encoding Technique

```python
data_cat.columns
```

```python
data_cat["cancellation"] = final["is_canceled"]
```

```python
def mean_encode(df, col, mean_col):
    df_dict = df.groupby([col])[mean_col].mean().to_dict()
    df[col] = df[col].map(df_dict)
    return df

for col in data_cat.columns[0:8]:
    data_cat = mean_encode(data_cat, col, "cancellation")
```

```python
data_cat.drop(["cancellation"] , axis = 1,inplace = True)
```

# Different algorithms 
# Naive Bayes
# Random Forest
# Decision Tree
# KNN

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier 
from sklearn.tree import DecisionTreeClassifier
```

```python
models = []

models.append(("Naive Bayes", GaussianNB()))
models.append(("Random Forest", RandomForestClassifier()))
models.append(("Decision Tree", DecisionTreeClassifier()))
models.append(("KNN", KNeighborsClassifier(n_neighbors = 5)))
```

```python
for name, model in models:
    print(name)
    model.fit(X_train, y_train)
 
    predictions = model.predict(X_test)
    
    from sklearn.metrics import confusion_matrix
    print(confusion_matrix(predictions, y_test))
    
    from sklearn.metrics import accuracy_score
    print(accuracy_score(predictions, y_test))
    
    print("\n")
```
